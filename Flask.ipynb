```python
from flask import Flask
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello world"

@app.route("/path")
def path():
    return "Hi Path"

@app.route("/churn_count")
def churn_count():
    df = pd.read_csv('Telco-Customer-Churn.csv')
    #result = list(df['Churn'].value_counts())
    result = df['customerID'].to_json()
    return result

@app.route("/get_customer_info/<customerID>", methods = ['GET'])
def get_customer_info(customerID):
    print(customerID)
    df = pd.read_csv('Telco-Customer-Churn.csv')
    result = df[df['customerID'] == customerID].to_json()
    return result

@app.route("/download_stock")
def download_stock():
    stock_symbol = 'AAPL'  # Thay 'AAPL' bằng mã chứng khoán bạn quan tâm
    stock_data = yf.download(stock_symbol, start='2022-01-01', end='2022-01-10')
    return stock_data.to_csv("final.csv")

@app.route("/get_stock")    
def get_stock():
    df = pd.read_csv('final.csv')
    df["Status"]=[inc_dec(c,o) for c, o in zip(df.Close,df.Open)]
    df["Middle"] = (df.Open+df.Close)/2
    df["Height"] = abs(df.Close-df.Open)
    fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

    fig.show()
    #return df.to_json()

def inc_dec(c, o):  #param ( c: Close , o: Open )
    if c > o:
        value="Increase"
    elif c < o:
        value="Decrease"
    else:
        value="Equal"
    return value

if __name__ == "__main__":
    app.run()

```
